# Compare Sequential Forgetting Full Experiment

Compare sequential_forgetting_full between `results` and `results_thesis`

```python
import json
from pathlib import Path

# Base paths
V1_PATH = Path("results")
V2_PATH = Path("results_thesis")
EXP_NAME = "sequential_forgetting_full"

def load_json(file_path):
    """Load JSON file, return None if not found."""
    try:
        if file_path.exists():
            return json.load(open(file_path))
    except:
        pass
    return None

print(f"Comparing {EXP_NAME}...")
```

```python
def compare_sequential_forgetting_full():
    """Compare sequential_forgetting_full experiment - direct all_seeds_summary.json comparison."""
    print(f"\n{EXP_NAME.upper()}")
    print("-" * 60)
    
    # Load both all_seeds_summary.json files directly
    v1_all = load_json(V1_PATH / EXP_NAME / "all_seeds_summary.json")
    v2_all = load_json(V2_PATH / EXP_NAME / "all_seeds_summary.json")
    
    print(f"\nV1 (results): {'OK' if v1_all else 'MISSING'}")
    print(f"V2 (results_thesis): {'OK' if v2_all else 'MISSING'}")
    
    if v1_all is None or v2_all is None:
        print("ERROR: Missing summary files")
        return
    
    print(f"\n=== V1 Summary ===")
    print(v1_all)
    
    print(f"\n=== V2 Summary ===")
    print(v2_all)
    
    # Compare
    print(f"\n=== COMPARISON ===")
    
    # Handle both dict and list formats
    if isinstance(v1_all, dict) and isinstance(v2_all, dict):
        all_keys = set(v1_all.keys()) | set(v2_all.keys())
        for key in sorted(all_keys):
            v1_val = v1_all.get(key)
            v2_val = v2_all.get(key)
            
            if isinstance(v1_val, (int, float)) and isinstance(v2_val, (int, float)):
                diff = v2_val - v1_val
                rel_diff = abs(diff / v1_val * 100) if v1_val != 0 else 0
                is_ok = abs(diff) < 0.01 or rel_diff < 2
                status = "OK" if is_ok else "DIFF"
                print(f"{key:20} V1: {v1_val:.6f}  V2: {v2_val:.6f}  Diff: {diff:+.6f} [{status}]")
            else:
                match = "=" if v1_val == v2_val else "!="
                print(f"{key:20} V1: {v1_val}  V2: {v2_val} [{match}]")
    else:
        print(f"V1 type: {type(v1_all)}")
        print(f"V2 type: {type(v2_all)}")
        print("Format check - files loaded successfully")

compare_sequential_forgetting_full()
```
